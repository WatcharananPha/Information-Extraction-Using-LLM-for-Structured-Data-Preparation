## Step 3: Entity Extraction

```python
def retry(n):
    """
    A decorator to retry a function n times upon failure, and print the number of attempts made.

    Parameters:
    -----------
    n : int
        The number of retry attempts.

    Returns:
    --------
    Callable
        A decorator that retries the decorated function up to n times.
    """
    def decorator_retry(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            attempts = 0
            while attempts < n:
                try:
                    attempts += 1
                    # Pass the attempt count to the decorated function
                    result = func(*args, attempt_number=attempts, **kwargs)
                    print(f"Success on attempt {attempts}")
                    return result
                except Exception as e:
                    print(f"Attempt {attempts} failed: {e}")
                    if attempts == n:
                        raise  # Re-raise the exception if it's the last attempt
        return wrapper
    return decorator_retry
```

```python
# Define the prompt for extracting named entities
@retry(n=3)
def extract_entities(text, attempt_number):
    prompt = f"""
    ### instruction
    Extract all named entities (ATTRACTION, LOCATION, PERSON, ORGANIZATION, ANTIQUE, CULTURE) with detailed descriptions
    from the following text. Format the output as a JSON array of entities.

    # Example Output Format
    [
        {{
            "name": "วัดศาลาแดง",
            "type": "ATTRACTION",
            "description": "วัดที่มีชื่อเสียงในพื้นที่ อยู่ติดกับเมืองเมาะตะมะ เป็นจุดสำคัญของชุมชนในอดีตและปัจจุบัน."
        }}
    ]

    ### input:
    {text}

    ### output:
    """
    res = model.generate_content(prompt)
    d = res.response.strip()
    parsed_data = json.loads(d)
    return parsed_data, attempt_number

def as_series(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return pd.Series(result)
    return wrapper

@as_series
def warp_extract_entities(x):
    parsed_data, attempt_number = extract_entities(x)
    return parsed_data, attempt_number
```

```python
# Process the input text to extract entities
input_df = text_df.copy()
input_df[['data', 'attempt']] = input_df.apply(lambda row: warp_extract_entities(row['text']), axis=1)
entity_output_df = input_df.copy()
entity_output_df.info()
entity_output_df.head()
```

```python
# Convert to dataframe
# Expand the 'data' column into separate rows
rows = []
chunk_ind = 0
for _, row in entity_output_df.iterrows():
    page = row['page']
    data_items = row['data']

    for item in data_items:
        rows.append({
            'page':page,
            'chunk': chunk_ind,
            'entity_name': item.get('name', None),
            'entity_type': item.get('type', None),
            'description': item.get('description', None)
        })

# Create a new DataFrame from the expanded rows
entity_df = pd.DataFrame(rows)
entity_df
```

```python
# Define the prompt for extracting events
@retry(n=3)
def extract_events(text, attempt_number):
    prompt = f"""
    ### instruction
    Extract all events with their descriptions and time information from the following text. Format the output as a JSON array of events.

    ### Example Output Format
    [
        {{
            "name": "วันอนุรักษ์มรดกไทย",
            "description": "วันสำคัญที่ประชาชนร่วมมือกันเพื่ออนุรักษ์มรดกทางศิลปวัฒนธรรมของชาติไทย.",
            "time": "2 เมษายน ทุกปี"
        }},
        {{
            "name": "งานเฉลิมฉลองเตาสามโคก",
            "description": "งานประจำปีที่จัดขึ้นเพื่อส่งเสริมการท่องเที่ยวและอนุรักษ์เครื่องปั้นดินเผาของเตาสามโคก.",
            "time": "เมษายน 2568"
        }}
    ]

    ### input:
    {text}

    ### output:
    """
    res = model.generate_content(prompt)
    d = res.response.strip()
    parsed_data = json.loads(d)
    return parsed_data, attempt_number

@as_series
def warp_extract_events(x):
    parsed_data, attempt_number = extract_events(x)
    return parsed_data, attempt_number
```

```python
# Process the input text to extract events
input_df = text_df.copy()
input_df[['data', 'attempt']] = input_df.apply(lambda row: warp_extract_events(row['text']), axis=1)
event_output_df = input_df.copy()
event_output_df.info()
event_output_df.head()
```
